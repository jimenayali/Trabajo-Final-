```python
import pandas as pd
import altair as alt
```

```python
happy_report_2021 = pd.read_csv("C:/Users/PROPIETARIO/Downloads/world-happiness-report-2021.csv")
```

```python
happy_report_2021
```

```python
happy_report_2021.columns= happy_report_2021.columns.str.lower()
happy_report_2021
```

```python
#Seria interesante ver que regiones del mundo son más felices y cuales son las menos felices
median_happiness = happy_report_2021['Ladder score'].median()
print(f"La mediana global de felicidad es: {median_happiness}")
```

```python
#la mediana de felicidad por región
regional_happiness_median = happy_report_2021.groupby('regional indicator')['ladder score'].median()
```

```python
#categorizar regiones en función de la mediana
def categorizar_regiones(happy_report_2021):
    '''
    Clasifica las regiones como "Más feliz" o "Menos feliz" en función de la mediana global de felicidad.
    
    Input: ladder score (float)
    Output: dict: Un diccionario donde las claves son las regiones y los valores son "Más feliz" o "Menos feliz".
    ''' 
    region_classification = {} 
    for region, score in regional_happiness_median.items():
        if score > median_happiness:
            region_classification[region] = 'Más feliz'
        else:
            region_classification[region] = 'Menos feliz'

    return region_classification
        
```

```python
region_classification = categorizar_regiones(happy_report_2021)
```

```python
happy_report_2021['regional_happiness'] = happy_report_2021['regional indicator'].map(region_classification)
```

```python
happy_report_2021
```

```python
alt.Chart(happy_report_2021).mark_bar().encode(
    x=alt.X("count():Q", title="Número de Países"),
    y=alt.Y("regional indicator:N", title="Región", sort="-x"),
    color=alt.Color("regional_happiness:N", title="Categoría de Felicidad"),
    tooltip=["regional indicator", "regional_happiness", "count()"]
).properties(
    title="Regiones Más Felices y Menos Felices",
    width=600,
    height=400
).interactive()
```

Africa Subsahariana es la región del mundo que tiene más países con un puntaje de felicidad menor a la mediana global. 
De los países con una mediana menor al puntaje global de felicidad, la menor cantidad se encuentra en Asia del Sur.

Europa Occidental es la región del mundo que tiene más países con un puntaje mayor a la mediana global. 
De los países con una mediana mayor al puntaje global de felicidad, la menor cantidad se encuentra en Norte America y Anz. 
 
Me pregunto si las variables expectativa de vida saludable, percepcion de la corrupción o libertad de expresión son las que varian más entre Africa Subsahariana y Europa Occidental. 

```python
africa_y_europa = happy_report_2021[happy_report_2021['regional indicator'].isin(['Sub-Saharan Africa', 'Western Europe'])]
africa_y_europa
```

```python
freedom = alt.Chart(africa_y_europa).mark_boxplot().encode(
    x=alt.X('regional indicator:N', title='Región'),
    y=alt.Y('freedom to make life choices:Q', title='Libertad para tomar decisiones'),
    color=alt.Color('regional indicator:N', title='Región')
    ).properties(
    title='Distribución de Libertad para Tomar Decisiones'
    )

freedom
```

```python
life_expectancy= alt.Chart(africa_y_europa).mark_boxplot().encode(
    x=alt.X('regional indicator:N', title='Región'),
    y=alt.Y('healthy life expectancy:Q', title='Expectativa de vida saludable'),
    color=alt.Color('regional indicator:N', title='Región')
    ).properties(
    title='Distribución de Expectativa de Vida Saludable'
    )

life_expectancy
```

```python
corruption = alt.Chart(africa_y_europa).mark_boxplot().encode(
    x=alt.X('regional indicator:N', title='Región'),
    y=alt.Y('perceptions of corruption:Q', title='Percepción de Corrupción'),
    color=alt.Color('regional indicator:N', title='Región')
    ).properties(
    title='Distribución de Percepción de Corrupción'
    )

corruption
```

```python
combined = alt.hconcat(freedom, life_expectancy, corruption)
combined
```
